## Karar Ağaçları İle Sınıflandırma
Karar ağaçları, verileri yüksek doğruluk dereceleriyle sınıflandırmanıza izin veren sezgisel, denetimli bir makine öğrenimi algoritmasıdır. Karar Ağaçları, karar verirken genellikle insanın düşünme yeteneğini taklit eder ve ağaç benzeri bir yapı gösterdiği için anlaşılması kolaydır.

Bir pazarlama yöneticisi olarak, ürününüzü satın alma olasılığı en yüksek olan bir dizi müşteri istiyorsunuz. Kitlenizi bularak pazarlama bütçenizden bu şekilde tasarruf edebilirsiniz. Bir kredi yöneticisi olarak, daha düşük bir kredi temerrüt oranı elde etmek için riskli kredi başvurularını belirlemeniz gerekir. Müşterileri potansiyel ve potansiyel olmayan müşteriler veya güvenli veya riskli kredi başvuruları grubuna ayırma işlemi, bir sınıflandırma problemi olarak bilinir. Sınıflandırma, **öğrenme adımı** ve **tahmin adımı** olmak üzere iki aşamalı bir süreçtir. *Öğrenme adımında*, verilen eğitim verilerine dayalı olarak model geliştirilir. *Tahmin adımında*, model verilen veri için yanıtı tahmin etmek için kullanılır. Karar Ağacı, anlaşılması ve yorumlanması en kolay ve popüler sınıflandırma algoritmalarından biridir. Hem sınıflandırma hem de regresyon türü problemler için kullanılabilir.

Karar Ağacı, beyaz kutu tipi bir makine öğrenimi algoritmasıdır. Neural Network gibi kara kutu tipi algoritmalarda bulunmayan dahili karar verme mantığını paylaşır. Eğitim süresi sinir ağı algoritmasına göre daha hızlıdır. Karar ağaçlarının zaman karmaşıklığı, verilen verideki kayıt sayısı ve öznitelik sayısının bir fonksiyonudur. Karar ağacı, olasılık dağılım varsayımlarına bağlı olmayan, dağılımdan bağımsız veya parametrik olmayan bir yöntemdir. Karar ağaçları, yüksek boyutlu verileri iyi bir doğrulukla işleyebilir.



### Karar Ağaçlarında Dallanma Kriterleri
Karar ağaçlarında en önemli sorunlardan birisi herhangi bir kökten itibaren dallanmanın hangi kıstasa göre yapılacağıdır.Her farklı kriter için bir karar ağacı algoritması karşılık gelmektedir. 

Algoritmaları gruplandıracak olursak;
* Entropiye Dayalı Algoritmalar
* Sınıflandırma ve Regresyon ağaçları
* Bellek tabanlı sınıflandırma algoritmaları

## Karar Ağacı Algoritması Nasıl Çalışır?
Herhangi bir karar ağacı algoritmasının arkasındaki temel fikir şu şekildedir:

1. Ağaca, tüm veri setini içeren kök düğümle başlayın, diyor S.
2. Nitelik Seçim Ölçütünü (ASM) kullanarak veri kümesindeki en iyi özelliği bulun .
3. S'yi en iyi nitelikler için olası değerleri içeren alt kümelere bölün.
4. En iyi özelliği içeren karar ağacı düğümünü oluşturun.
5. Adım -3'te oluşturulan veri setinin alt kümelerini kullanarak yinelemeli olarak yeni karar ağaçları oluşturun. Düğümleri daha fazla sınıflandıramayacağınız ve son düğümü yaprak düğüm olarak adlandıracağınız bir aşamaya gelene kadar bu işleme devam edin.

### Ağaç her düzeyde hangi değişkenin dallanacağına nasıl karar verir?

* Açgözlü yukarıdan aşağıya yaklaşım

Karar ağaçları, özyinelemeli ikili bölme olarak bilinen yukarıdan aşağıya açgözlü bir yaklaşımı izler. Özyinelemeli ikili bölme yaklaşımı yukarıdan aşağıyadır çünkü ağacın tepesinden başlar ve ardından tahmin uzayını art arda böler. Her bölmede tahmin alanı 2'ye bölünür ve aşağıyı gösteren iki yeni dalla gösterilir. Algoritmaya açgözlü denir, çünkü sürecin her adımında, o adım için en iyi ayrım yapılır. İleriye doğru projeksiyon yapmaz ve ağacın tamamı için daha uygun olabilecek bir bölünme seçmeye çalışır.

Bu nedenle algoritma, tüm değişkenleri bazı istatistiksel kriterlere göre değerlendirir ve ardından kriterler üzerinde en iyi performansı gösteren değişkeni seçer.

## Değişken seçim kriteri

Nitelik seçim ölçüsü, verileri mümkün olan en iyi şekilde bölümleyen bölme kriterini seçmek için bir buluşsal yöntemdir. Belirli bir düğümdeki demetler için kesme noktalarını belirlememize yardımcı olduğu için bölme kuralları olarak da bilinir. ASM, verilen veri setini açıklayarak her özelliğe (veya özniteliğe) bir sıralama sağlar. En iyi puan özelliği, bir bölme özelliği olarak seçilecektir . Sürekli değerli bir öznitelik olması durumunda, dallar için ayrık noktaların da tanımlanması gerekir. 

Kararın gerçek karmaşıklığı ve karmaşıklığı burada yatmaktadır. Değişkenler, her karar düğümünde uygulanan karmaşık bir istatistiksel kritere göre seçilir. Artık Karar Ağaçlarında değişken seçim kriteri iki yaklaşımla yapılabilir:

1. Entropi ve Bilgi Kazanımı

2. Gini Endeksi

Her iki kriter de genel olarak benzerdir ve hangi değişkenin verileri bölerek altta yatan alt düğümlerin en homojen veya saf olmasına yol açacağını belirlemeye çalışır. Her ikisi de farklı Karar Ağacı algoritmalarında kullanılır. Karışıklığa ek olarak, hangisinin tercih edilen yaklaşım olduğu açık değildir. Yani, her ikisini de anlamak gerekir.

> ### Bilgi Kazanımı 
> Bilgi kazancı, bir özniteliğe dayalı olarak bir veri kümesinin bölümlenmesinden sonra entropideki değişikliklerin ölçümüdür. Bir özelliğin bize bir sınıf hakkında ne kadar bilgi sağladığını hesaplar. Bilgi kazancının değerine göre düğümü bölüp karar ağacını oluşturuyoruz.Bir karar ağacı algoritması her zaman bilgi kazancının değerini maksimize etmeye çalışır ve en yüksek bilgi kazancına sahip bir düğüm/öznitelik önce bölünür.

> Shannon, girdi setinin safsızlığını ölçen entropi kavramını icat etmiştir. **Fizik ve matematikte entropi, sistemdeki rastgelelik veya kirlilik olarak adlandırılır. Bilgi teorisinde, bir grup örnekteki safsızlığı ifade eder.** Bilgi kazancı entropinin azalmasıdır. Bilgi kazancı, verilen öznitelik değerlerine dayalı olarak veri kümesinin bölünmesinden önceki entropi ile bölünmesinden sonraki ortalama entropi arasındaki farkı hesaplar. ID3 (Iterative Dichotomiser) karar ağacı algoritması bilgi kazancını kullanır.

> <div class="alert alert-block alert-success">
<b>ID3 Algoritması :</b>
ID3 algoritması, entropi’den faydalanarak sınıflandırma problemlerinin çözülmesine imkan tanır. ID3 algoritmasında sistemin genel entropisinin hesaplanmasının ardından her bir özniteliğin entropisi de ayrı ayrı hesaplanır. Hesaplanan entropi değerleri, genel entropiden çıkartılarak; hangi özniteliğin entropi değerini en çok azalttığı tespit edilir.
</div>

### Kazanç Oranı
Bilgi kazanımı, birçok sonucu olan özellik için önyargılıdır. Bu, çok sayıda farklı değere sahip özelliği tercih ettiği anlamına gelir. Örneğin, müşteri_kimliği gibi benzersiz bir tanıtıcıya sahip bir özniteliğin saf bölümleme nedeniyle sıfır bilgi(D) içerdiğini düşünün. Bu, bilgi kazancını en üst düzeye çıkarır ve yararsız bölümleme oluşturur.

ID3'ün geliştirilmiş hali olan C4.5, kazanç oranı olarak bilinen bir bilgi kazancı uzantısını kullanır. Kazanç oranı, Bölünmüş Bilgiyi kullanarak bilgi kazancını normalleştirerek yanlılık sorununu çözer. 
> 

> T veri tabanı X testine göre bölmekle elde edilen bilgileri ölçmek için **’’kazanç ölçütü’’**d adı verilen bir ifadeye başvurulur. Bu ölçüt; <br>

> ### Gini Endeksi:
Bir karar ağacını bölmenin diğer yolu Gini İndeksi'dir. Entropi ve Bilgi Kazanımı yöntemi, bir düğümdeki saflık ve safsızlık üzerine odaklanır. Gini Endeksi veya Safsızlık, rastgele seçildiğinde rastgele bir örneğin yanlış sınıflandırılma olasılığını ölçer. Gini Endeksi ne kadar düşükse, yanlış sınıflandırma olasılığı o kadar düşük olur.
> * Gini İndeksi formülü
> 
> ### Gini Endeksi ve Bilgi Kazanımı
> Hangi kirlilik ölçümünün kullanıldığına bağlı olarak, ağaç sınıflandırma sonuçları değişebilir. Bu, modeliniz üzerinde küçük (veya bazen büyük) bir etki yapabilir. Farklı Karar Ağacı algoritmaları tarafından tercih edilen bir yaklaşım yok gibi görünüyor. Örneğin CART, Gini'yi kullanır; ID3 ve C4.5 Entropy kullanır.
> Gini indeksinin maksimum safsızlığı 0,5 ve maksimum saflığı 0'dır, oysa Entropy'nin maksimum safsızlığı 1 ve maksimum saflığı 0'dır.
> ### Gini Entropiye Karşı
Gini Endeksi ve Entropinin iki temel farkı vardır:
> Gini Endeksi [0, 0.5] aralığında değerler alırken, Entropi aralığı [0, 1] aralığındadır. Aşağıdaki şekilde, her ikisi de temsil edilmektedir. Aralarındaki çok anlamlı olmayan farkları somut olarak görebilmek için gini indeksi de iki ile çarpılarak temsil edilmiştir.
> 
> * Hesaplama açısından entropi, logaritmalardan yararlandığı için daha karmaşıktır ve sonuç olarak Gini Endeksinin hesaplanması daha hızlı olacaktır.

Yukarıdaki görüntüye dayanarak, verilerde bir takım net ayrımlar olduğunu görebiliriz. Bu, bir karar ağacı sınıflandırıcısı kullanarak verilerimizi saf bölmelere ayırmada oldukça yardımcı olabilir.

> Şimdi sayısal olmayan sütunları bırakarak sayısal değerleri ele alalım.

```python
data = pd.read_csv(
    'https://github.com/datagy/data/raw/main/titanic.csv', 
    usecols=['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare'])
data = data.dropna()
```

Yukarıdaki kodda, 'Sex' ve 'Embarked' sütunlarını kaldırarak sadece sayısal sütunlarının olduğu bir data oluşturduk. Şimdi verilerimizi X ve y olmak üzere iki parçaya ayıralım.

```python
X = data.copy()  
# Özellikler matrisimiz. Bir matris olduğu için büyük harf ile ifade ediyoruz.
y = X.pop('Survived')  
# Titanikte hayatta kalanları hedef değişkenimiz olarak ayarlıyoruz.
```

> **pop()** fonksiyonu ilgili değeri çıkarır ve geriye çıkarttığı değeri döndürür.

Verilerimizi Sklearn kütüphanesini kullanarak **Eğitim** ve **Test** verilerine bölelim.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 100)
```

## Sklearn'de DecisionTreeClassifier'ı Anlamak

> DecisionTreeClassifierSklearn'de sınıfta bulunan parametreler

```python
DecisionTreeClassifier(
    criterion = 'gini', # Bir bölmenin kalitesini ölçme işlevi. 'gini' - 'entropy'.
    splitter = 'best', # En iyi bölünmeyi seçme stratejisi. 'best' - 'random'
    max_depth = None, # Ağacın maksimum derinliği. none ise, düğümler tüm yapraklar saf olana veya min_samples_split'ten daha azını içerene kadar genişletilir.
    min_samples_split = 2, # Bir düğümü bölmek için gereken minimum örnek sayısı.
    min_samples_leaf = 1, # Bir yaprak düğümde olması gereken minimum örnek sayısı.
    min_weight_fraction_leaf = 0.0, # Bir düğümde olması gereken tüm girdi örneklerinin ağırlıklarının toplamının minimum ağırlıklı fraksiyonu.
    max_features = None, # En iyi ayrımı ararken dikkate alınması gereken özelliklerin sayısı. "int-float-'auto'-'sqrt'-'log2'-none" olabilir.
    random_state = None, # Tahmin edicinin rastgeleliği için kontrol
    max_leaf_nodes=None, # Maksimum sayıda düğüm içeren bir ağaç büyütün. Yok ise, sınırsız sayıda mümkündür.
    min_impurity_decrease=0.0, # Bu bölme, safsızlığı bu değerden büyük veya bu değere eşit azaltırsa, bir düğüm bölünecektir.
    class_weight=None, # Farklı sınıflarla ilişkili ağırlıklar.
    ccp_alpha=0.0 # Minimum Maliyet-Karmaşıklık Budaması için kullanılan karmaşıklık parametresi.
)
```

Bu eğitimde, kapsamını korumak için aşağıdaki parametrelere odaklanacağız: **criterion, max_depth, max_features, splitter**. 

Sklearn ile ilgili harika şeylerden biri, model oluşturmanın ardındaki karmaşıklığın çoğunu soyutlama yeteneğidir. Bu nedenle, kendimiz herhangi bir karar vermeden aslında bir Karar Ağacı oluşturabiliriz. Bunu, sınıf tarafından sağlanan varsayılan parametreleri kullanarak yapabiliriz.

* Şimdi, Sklearn'ü kullanarak ilk karar ağacı sınıflandırıcımızı nasıl oluşturabileceğimize bakalım!

```python
clf = DecisionTreeClassifier() # modelimizi oluşturduk
clf.fit(X_train, y_train) # modelimize eğitim verilerimiz aktardık.
```

Scikit-Learn bizim yerimize (daha iyi veya daha kötü!) tüm kararları vermekle ilgilenir. Şimdi yeni oluşturduğumuz bu model ile nasıl tahmin yapabileceğimize bakalım:

```python
predictions = clf.predict(X_test)
predictions[:5]
```

Öngörülen değerlerimizin ilk beş kaydını yazdırdığımızda, burada ki 0 değeri yolcunun hayatta kalmadığını, 1 ise hayatta kaldığını gösterir.

### Python's Sklearn'de Karar Ağacı Sınıflandırıcı Algoritmasını Doğrulama

Farklı makine öğrenimi modelleri türleri, farklı doğruluk ölçümlerine dayanır. X_test dizisini kullanarak tahminler yaptığımızda, sklearn bir dizi tahmin döndürdü. Bunların gerçek değerlerini zaten biliyoruz: y_test'te depolanırlar.

Algoritmanın etkinliğini ölçen 1 üzerinden bir orantı döndürmek için accuracy_score() sklearn işlevini kullanabiliriz. Doğruluk puanı, tüm tahminlerin toplamından doğru tahminlerin oranına bakar.


```python
accuracy_score(y_test, predictions)
```

Döndürülen doğruluk puanının 0,68 veya %68 olduğunu görebiliriz. Bu, tahmin etmeye çalıştığımız tüm değerlerin %68'inin doğru olduğu anlamına gelir.

# Sklearn'de Karar Ağacı Sınıflandırıcıları için Hiperparametre Ayarı

Bu öğreticiyi bitirmek için, bazı hiper parametrelerini ayarlayarak modelimizin doğruluğunu nasıl geliştirebileceğimize bir göz atalım. Hiper parametreler, bir makine öğrenimi modeli oluştururken belirttiğiniz değişkenlerdir. Bu, örneğin, algoritmanın verileri nasıl böldüğünü (entropi veya gini safsızlığı ile) içerir.

Hiper parametre ayarı, daha sonra, daha yüksek bir doğruluk puanı sağlamak için bu değerleri ayarlama sürecini ifade eder. Bunu yapmanın bir yolu, basitçe, farklı değerleri girmek ve hangi hiper parametrelerin en yüksek puanı döndürdüğünü görmektir.

Scikit-learn, bir dizi farklı kombinasyonu denemek yerine, bu testleri otomatikleştirmenin bir yolunu sunar. Bu yöntem, işlemi önemli ölçüde daha hızlı hale getiren GridSearchCV yöntemidir. Denemek için farklı değerlerden oluşan bir sözlük sağlamamız yeterlidir ve Scikit-Learn süreci bizim yerimize halleder.

Ayrıca, sınıf bir çapraz doğrulama(cross-validation) sürecini tamamlar. Bu, sınıfın fazla uydurmayı önlemeye yardımcı olmak için farklı eğitim ve test verileri kombinasyonları arasında geçiş yapacağı anlamına gelir. Örneğin, %20'lik bir test boyutu belirlediğimizde, çapraz doğrulama bütüne göre bu %20'nin farklı bölümleri arasında geçiş yapacaktır.

> Hem en iyi hiperparametreleri bulmak hem de aynı anda çapraz doğrulama uygulamak için GridSearchCV sınıfını nasıl uygulayabileceğimizi görelim.

Bunu yapabilmek için öncelikle hangi hiperparametreleri test edeceğimize karar vermemiz gerekiyor.

```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],  # bir bölmenin kalitesini belirlemek için kullanılan işlev
    'max_depth':  [None, 2, 4, 6, 8, 10], # ağacın maksimum derinliği
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8], # bölme yaparken dikkate alınması gereken maksimum özellik sayısı
    'splitter': ['best', 'random'] # her düğümde bölmeyi seçmek için kullanılan strateji
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)

clf.fit(X_train, y_train)
print(clf.best_params_)
```

* Bu parametreler en iyi parametreler olarak etiketlense de, bunun aktardığımız parametre kombinasyonları bağlamında olduğunu unutmayın. Teorik olarak daha iyi parametreler olabilir.

Şimdi modelimizi yeniden oluşturup değerlendirelim ve şimdi skorun ne olduğunu görelim:

```python
X = pd.read_csv(
    'https://github.com/datagy/data/raw/main/titanic.csv', 
    usecols=['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex', 'Embarked'])
X = X.dropna()
y = X.pop('Survived')

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 100)

column_transformer = make_column_transformer(
    (OneHotEncoder(), ['Sex', 'Embarked']),
    remainder='passthrough')

X_train = column_transformer.fit_transform(X_train)
X_train = pd.DataFrame(data=X_train, columns=column_transformer.get_feature_names())

X_test = column_transformer.transform(X_test)
X_test = pd.DataFrame(data=X_test, columns=column_transformer.get_feature_names())

clf = DecisionTreeClassifier(max_depth=4, criterion='entropy', max_features=0.6, splitter='best')
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)

print(accuracy_score(y_test, predictions))
```

Yukarıdaki kod bloğu, modelimizin %76'nın üzerinde doğruluk sağladığı kodun tamamını içerir!
