Test kmesindeki bir deerin eitim kmesinde gzlemlenemeyen bir deeri varsa olaslk deeri olarak 0 verir yani tahmin yapamaz. Bu durum genellikle Zero Frequency ( Sfr Frekans ) adyla bilinir. Bu durumu zmek iin dzeltme teknikleri kullanlabilir. En basit dzeltme tekniklerinden biri Laplace tahmini olarak bilinir.

> P(Evet|Gneli) = P(Gneli|Evet) x P(Evet) / P(Gneli)

>
> P(Gneli) = 0.35
>
> P(Evet) = 0,71
>
> Yani P(Evet|Gneli) = 0,3 x 0,71 / 0,35 = 0,60
>
> P(Hayr|Gneli) = P(Gneli|Hayr) x P(Hayr) / P(Gneli)
>
> P(Gneli|HAYIR) = 2/4 = 0,5
>
> P(Hayr) = 0.29
>
> P(Gneli) = 0.35
>
> Yani P(Hayr|Gneli)= 0,5 x 0,29 / 0,35 = 0,41
>
> Yukardaki hesaplamadan da grebileceimiz gibi, P(Evet|Gneli) > P(Hayr|Gneli)

```python
df = pd.read_excel('titanic.xlsx')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns', inplace=True)
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs.head(3)
```

```python
target[:3]
```

```python
# inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

Kukla deiken tuzak teorisi nedeniyle erkek stununu da brakyorum. Erkek ve diiyi temsil etmek iin bir stun yeterlidir

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.info()
```

Age colonunda eksik veri grnyor.

```python
inputs.columns[inputs.isna().any()] # eksik veri bulunan kolonlar bu ekilde de grebiliriz.
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
