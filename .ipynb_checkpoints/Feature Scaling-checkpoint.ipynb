```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({"WEIGHT":[15,18,12,10],
                   "PRICE":[1,3,2,5]},
                   index=['Orange','Apple','Banana','Grape'])
df
```

##  Min-Max Scaler


Verinin 0 ile 1 arasnda deerler ald bir durumdur. Burada dalm, verinin dalm ile benzerdir. Burada ar byk deerler yani outlier denilen  dta kalan verilere kar hassasiyet durumu vardr, bu yzden bu deerlerin fazla olduu bir durumda iyi bir performans gsteremeyebilir.

Dier yntemlerin aksine feature_range(min,max) parametresini kullanarak deikenlerin leklenecei aral kendimiz belirtebiliriz. Eer bu parametreyi bo brakrsak default olarak deikenleri [0,1] aralnda leklendirir. Negatif deerler varsa -1 ila 1 aralndaki verileri kltr. Aral [0,1] veya [0,5] veya [-1,1] gibi ayarlayabiliriz.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

print(scaler.fit_transform(df)) # 0 - 1 aralnda dntrlm array kts elde ediyoruz.

# Colon ve index deerlerini vererek Dataframe'e dntrelim.
df1 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

# leklenmemi verisetinin grafii
ax = df.plot.scatter(x = 'WEIGHT', 
                     y = 'PRICE', 
                     color = ['red','green','blue','orange'], # (c - color)
                     marker = '*',  # iareti
                     s = 80,  # iaretinin boyutu 
                     label = 'BEFORE SCALING')

# leklenmi ve leklenmemi verileri karlatrma grafii
df1.plot.scatter(x = 'WEIGHT', 
                 y = 'PRICE', 
                 color = ['red','green','blue','orange'],
                 marker = 'o', 
                 s = 60,
                 label = 'AFTER SCALING',
                 ax = ax) # ax ile 2 grafii birletiriyoruz.

plt.axhline(y = 0, color = 'red', alpha = 0.2) # axhline:Eksenler boyunca yatay bir izgi ekler.
plt.axvline(x = 0, color = 'red', alpha = 0.2) # axvline:Eksenler boyunca dikey bir izgi ekler.

# alpha:iaretinin effafl ayarlanr.
```

## Standard Scaler



Standardizasyon ileminde her bir stundaki gzlem birimi ortalamadan karlr ve standart sapmaya blnr. Bylelikle deikenler ortalamas 0, varyans 1 olan normal dalm aralna getirilmi olur. Ancak bu yntem verinin normal dald varsaymna dayanr ileme gemeden nce bu varsaym kontrol etmelisiniz. Veriler normal bir ekilde datlmyorsa, kullanlacak en iyi lekleyici bu deildir. Standardizasyon leklendirmesi veride bir bilgi kaybna/deiimine neden olmaz yalnzca deikenleri belirli bir standarda tam olur. Bu dnmden sonra deikenler [-1,1] arasnda deerler alrlar.


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

print(scaler.fit_transform(df)) # -1 ile 1 aralnda dntrlm array kts elde ediyoruz.

df2 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape']) 

ax = df.plot.scatter(x='WEIGHT',
                     y='PRICE',
                     color=['red','green','blue','orange'], 
                     marker = '*',
                     s=80, 
                     label='BREFORE SCALING')

df2.plot.scatter(x='WEIGHT',
                 y='PRICE',
                 color=['red','green','blue','orange'], 
                 marker = 'o',
                 s=80, 
                 label='AFTER SCALING',
                 ax = ax)

plt.axhline(y = 0, color='red', alpha=0.2)
plt.axvline(x = 0, color='red', alpha=0.2)
```

## Max Abs Scaler


Eitim setindeki her zelliin maksimum mutlak deeri 1.0 olacak ekilde her bir zellii ayr ayr lekler ve evirir. Verileri kaydrmaz/merkezlemez ve dolaysyla herhangi bir seyreklii yok etmez.
Yalnzca pozitif verilerde, bu lekleyici, Min Maks lekleyiciye benzer ekilde davranr ve bu nedenle, nemli aykr deerlerin varlndan da zarar grr.

Bu lekleyici, seyrek CSR veya CSC matrislerine de uygulanabilir.

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()

print(scaler.fit_transform(df)) 

df4 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

ax = df.plot.scatter(x='WEIGHT', 
                     y='PRICE',
                     color=['red','green','blue','yellow'], 
                     marker = '*',
                     s=80, 
                     label='BREFORE SCALING')

df4.plot.scatter(x='WEIGHT', 
                 y='PRICE', 
                 color=['red','green','blue','yellow'],
                 marker = 'o',
                 s=60,label='AFTER SCALING', 
                 ax = ax)

plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```

## Robust Scaler

Verilerimiz ok sayda aykr deer ieriyorsa verilerin ortalamasn ve standart sapmasn kullanarak leklendirme iyi sonu vermez. Dier leklendirme yntemleri minimum-maksimum- ortalama deeri kulland iin aykr deerlere duyarllard ancak Robust Scaler yntemi leklendirmede IQR (eyrekler Aklk) kulland iin aykr deerlere kar salam bir yntemdir.



Bu lekleyici medyan kaldrr ve verileri nicelik aralna gre leklendirir (varsaylan IQR: eyrekler Aras Aralk'tr). IQR, 1. eyrek (25. eyrek) ile 3. eyrek (75. eyrek) arasndaki aralktr. Bu lekleyicinin merkezleme ve leklendirme istatistikleri yzdelik dilimlere dayaldr ve bu nedenle birka byk marjinal aykr deerden etkilenmez. Aykr deerlerin dntrlen verilerde hala mevcut olduuna dikkat edin. Ayr bir aykr deer krplmas isteniyorsa, dorusal olmayan bir dnm gereklidir.

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

print(scaler.fit_transform(df)) 

df3 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

ax = df.plot.scatter(x='WEIGHT', 
                     y='PRICE',
                     color=['red','green','blue','yellow'], 
                     marker = '*',
                     s=80, 
                     label='BREFORE SCALING')

df3.plot.scatter(x='WEIGHT', 
                 y='PRICE', 
                 color=['red','green','blue','yellow'],
                 marker = 'o',
                 s=60,label='AFTER SCALING', 
                 ax = ax)

plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```

imdi bir aykr deer ekleyelim ve Standart Scaler ve Robust Scaler kullanarak leklendirmenin etkisinin (daire aykr deeri gsterir) ne olacan grelim.

```python
dfr = pd.DataFrame({'WEIGHT': [15, 18, 12,10,50],
                    'PRICE': [1,3,2,5,20]},
                    index = ['Orange','Apple','Banana','Grape','Jackfruit'])
print(dfr)
```

```python
# STANDARD SCALER

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df21 = pd.DataFrame(scaler.fit_transform(dfr),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape','Jackfruit'])

ax = dfr.plot.scatter(x='WEIGHT', 
                      y='PRICE',
                      color=['red','green','blue','orange','black'], 
                      marker = '*',
                      s=80, 
                      label='BREFORE SCALING');

df21.plot.scatter(x='WEIGHT', 
                  y='PRICE', 
                  color=['red','green','blue','orange','black'],
                  marker = 'o',
                  s=60,label='STANDARD', 
                  ax = ax,
                  figsize=(12,6))


# ROBUST SCALER

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

df31 = pd.DataFrame(scaler.fit_transform(dfr),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape','Jackfruit'])

df31.plot.scatter(x='WEIGHT', 
                  y='PRICE', 
                  color=['red','green','blue','orange','black'],
                  marker = 'v',
                  s=60,
                  label='ROBUST', 
                  ax = ax,
                  figsize=(12,6))

plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```

## Quantile Transformer Scaler

Bu yntem, zellikleri tek tip veya normal bir dalm izleyecek ekilde dntrr . Bu nedenle, belirli bir zellik iin, bu dnm en sk deerleri yayma eilimindedir. Ayn zamanda (marjinal) aykr deerlerin etkisini de azaltr: bu nedenle bu, salam bir n ileme plandr.

Bir zelliin kmlatif dalm ilevi, orijinal deerleri yanstmak iin kullanlr. Bu dnmn dorusal olmadna ve ayn lekte llen deikenler arasndaki dorusal korelasyonlar bozabileceine, ancak farkl leklerde llen deikenleri daha dorudan karlatrlabilir hale getirdiine dikkat edin. Bu bazen Rank scaler olarak da adlandrlr.

```python
from sklearn.preprocessing import QuantileTransformer
scaler = QuantileTransformer()

print(scaler.fit_transform(df))

df6 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

ax = df.plot.scatter(x='WEIGHT', 
                     y='PRICE',
                     color=['red','green','blue','yellow'], 
                     marker = '*',
                     s=80, 
                     label='BREFORE SCALING')

df6.plot.scatter(x='WEIGHT', 
                 y='PRICE', 
                 color=['red','green','blue','yellow'],
                 marker = 'o',
                 s=60,
                 label='AFTER SCALING', 
                 ax = ax,
                 figsize=(6,4))

plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```


> **!!!** Yukardaki rnek sadece aklama amaldr, nk Quantile dntrc, genellikle 1000'den fazla birok veri noktasna sahip byk bir veri setimiz olduunda kullanldr.

## Power Transformer Scaler
Power transformer dnm zellikle normalliin istendii durumlarda kullanlr. Deiken varyans durumunda varyans stabilize etmek ve verideki arpkl azaltmak iin maksimum olabilirlik yntemi kullanlr.

G transformatr, verileri daha fazla Gauss benzeri yapmak iin uygulanan bir parametrik, monoton dnmler ailesidir . Bu, aralk boyunca eit olmayan bir deikenin deikenlii (deiken varyans) veya normalliin istendii durumlar ile ilgili sorunlar modellemek iin kullanldr.

G dnm, varyans dengelemede ve maksimum olaslk tahmini yoluyla arpkl en aza indirmede en uygun leklendirme faktrn bulur. u anda PowerTransformer'n Sklearn uygulamas, Box-Cox dnmn ve Yeo-Johnson dnmn desteklemektedir. Varyans dengelemek ve arpkl en aza indirmek iin en uygun parametre, maksimum olaslk yoluyla tahmin edilir. Box-Cox, girdi verilerinin kesinlikle pozitif olmasn gerektirirken, Yeo-Johnson hem pozitif hem de negatif verileri destekler.

```python
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer(method='yeo-johnson')

print(scaler.fit_transform(df))

df5 = pd.DataFrame(scaler.fit_transform(df),
                   columns=['WEIGHT','PRICE'],
                   index = ['Orange','Apple','Banana','Grape'])

ax = df.plot.scatter(x='WEIGHT', 
                     y='PRICE',
                     color=['red','green','blue','yellow'], 
                     marker = '*',
                     s=80, 
                     label='BREFORE SCALING')

df5.plot.scatter(x='WEIGHT', 
                 y='PRICE', 
                 color=['red','green','blue','yellow'],
                 marker = 'o',
                 s=60,
                 label='AFTER SCALING', 
                 ax = ax)

plt.axhline(0, color='red',alpha=0.2)
plt.axvline(0, color='red',alpha=0.2)
```

## Unit Vector Scaler


lekleme, tm znitelik vektrnn birim uzunlukta olduu dikkate alnarak yaplr. Bu genellikle her bileeni vektrn klid uzunluuna (L2 Normu) blmek anlamna gelir. Baz uygulamalarda (rnein, histogram zellikleri), zellik vektrnn L1 normunu kullanmak daha pratik olabilir.

Min-Maks leklendirme gibi, Birim Vektr teknii [0,1] aralnda deerler retir. Kesin snrlar olan zelliklerle urarken bu olduka kullanldr. rnein, grnt verileriyle urarken renkler yalnzca 0 ile 255 arasnda deiebilir.

```python
# L1 normlu birim vektr
df8 = df.apply(lambda x : x / np.linalg.norm(x,1))
df8
```

```python
# L2 normlu birim vektr
df9 = df.apply(lambda x : x / np.linalg.norm(x,2))
df9
```
