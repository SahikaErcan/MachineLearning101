### Gradient Nedir?
Makine öğreniminde gradyan, birden fazla girdi değişkeni olan bir fonksiyonun türevidir. Matematiksel terimlerle bir fonksiyonun eğimi olarak bilinen gradyan, basitçe hatadaki değişime göre tüm ağırlıklardaki değişimi ölçer.
* Yol boyunca mümkün olan en az adımla bir tepenin zirvesine tırmanmak isteyen gözleri bağlı bir adam düşünün. Tepeye yakın olmadığı sürece yapabileceği en dik yönde gerçekten büyük adımlar atarak tepeye tırmanmaya başlayabilir. Ancak zirveye yaklaştıkça adımları gittikçe küçülür. Bu süreç, gradyan kullanılarak matematiksel olarak açıklanabilir.
    Aşağıdaki görselin tepemizi yukarıdan gösterdiğini ve kırmızı okların tırmanıcımızın basamakları olduğunu hayal edin. Bu bağlamda bir eğimi, gözü kapalı bir adamın atabileceği en dik adımın yönünü ve bu adımın ne kadar sürmesi gerektiğini içeren bir vektör olarak düşünün.

     

    X0'dan X1'e uzanan gradyanın, X3'ten X4'e ulaşandan çok daha uzun olduğuna dikkat edin. Bunun nedeni, vektörün uzunluğunu belirleyen tepenin dikliğinin/eğiminin daha az olmasıdır. Bu, tepe örneğini mükemmel bir şekilde temsil eder çünkü tepe tırmandıkça daha az dikleşir. Bu nedenle, azaltılmış bir eğim, yokuş tırmanıcı için azaltılmış bir eğim ve azaltılmış bir adım boyutu ile birlikte gider.
  
### Gradient Descent Nasıl Çalışır?
Gradient Descent Algorithm, geçerli konumdaki gradyanı kullanarak bir sonraki noktayı yinelemeli olarak hesaplar, bunu ölçeklendirir (bir öğrenme hızına göre) ve elde edilen değeri mevcut konumdan çıkarır (bir adım atar). Değeri çıkarır, çünkü işlevi en aza indirmek istiyoruz (maksimumlaştırmak için toplama olacaktır). Bu işlem şu şekilde yazılabilir:



Gradyanı ölçeklendiren ve dolayısıyla adım boyutunu kontrol eden önemli bir η parametresi vardır . Makine öğreniminde buna öğrenme oranı denir ve performans üzerinde güçlü bir etkiye sahiptir.

Bir tepeye tırmanmak yerine, yokuş aşağı inişi bir vadinin dibine inmek olarak düşünün.

Aşağıdaki denklem, gradyan iniş algoritmasının ne yaptığını açıklar: 

    b: Tırmanıcımızın bir sonraki konumudur, 
    a: Mevcut konumunu temsil eder. 
    Eksi işareti: Gradient descent algoritmasının küçültme kısmını ifade eder. 
    Gama (ɣ): Bir bekleme faktörüdür. 
    Gradyan terimi ( Δf(a) ): Basitçe en dik inişin yönüdür.
    
   

Bu formül temel olarak bize gitmemiz gereken bir sonraki konumu, yani en dik inişin yönünü söylüyor.    

* Bir makine öğrenme probleminiz olduğunu ve maliyet fonksiyonunuzu J ( w , b ) en aza indirmek ve parametrelerini ( w ve b ) değiştirerek yerel minimumuna ulaşmak için algoritmanızı gradyan inişiyle eğitmek istediğinizi hayal edin.
    Aşağıdaki görüntü, parametreleri ( w ve b ) temsil eden yatay eksenleri gösterirken, dikey eksenlerde maliyet fonksiyonu J ( w , b ) temsil edilir. Gradyan iniş dışbükey bir fonksiyondur.

    

    Maliyet fonksiyonunun minimumuna (kırmızı okla işaretlenmiş) karşılık gelen w  ve b değerlerini bulmak istediğimizi biliyoruz . Doğru değerleri bulmaya başlamak için w  ve b'yi  bazı rasgele sayılarla başlatıyoruz. Gradyan iniş daha sonra bu noktada başlar (resmimizin üst kısmında bir yerde) ve en dik aşağı yönde (yani, resmin üstünden altına) adım adım ilerler ve maliyetin olduğu noktaya ulaşır. Fonksiyon mümkün olduğu kadar küçüktür.
    
### Öğrenme Hızının Önemi
Gradyan inişinin yerel minimum yönünde attığı adımların ne kadar büyük olduğu, optimal ağırlıklara doğru ne kadar hızlı veya yavaş ilerleyeceğimizi belirleyen öğrenme oranı tarafından belirlenir.

Gradyan iniş algoritmasının yerel minimuma ulaşması için, öğrenme oranını ne çok düşük ne de çok yüksek olan uygun bir değere ayarlamalıyız. Bu önemlidir, çünkü attığı adımlar çok büyükse, yerel minimuma ulaşamayabilir çünkü gradyan inişinin dışbükey işlevi arasında ileri geri sıçrar. Öğrenme oranını çok küçük bir değere ayarlarsak, kademeli iniş sonunda yerel minimuma ulaşacaktır, ancak bu biraz zaman alabilir. 



Bu nedenle öğrenme oranı asla çok yüksek veya çok düşük olmamalıdır.

### Düzgün Çalıştığından Nasıl Emin Olabilirsiniz?
Gradyan iniş algoritmasının düzgün çalıştığından emin olmanın iyi bir yolu, optimizasyon çalışırken maliyet fonksiyonunu çizmektir. Yineleme sayısını x eksenine ve maliyet fonksiyonunun değerini y eksenine koyun. Bu, her kademeli iniş yinelemesinden sonra maliyet fonksiyonunuzun değerini görmenize yardımcı olur ve öğrenme oranınızın ne kadar uygun olduğunu kolayca belirlemenizi sağlar. Bunun için farklı değerler deneyebilir ve hepsini bir araya getirebilirsiniz. Aşağıdaki soldaki resim böyle bir çizimi gösterirken, sağdaki resim iyi ve kötü öğrenme oranları arasındaki farkı gösteriyor.


Gradyan iniş algoritması düzgün çalışıyorsa, maliyet fonksiyonu her yinelemeden sonra azalmalıdır.

Gradyan iniş, maliyet fonksiyonunu artık azaltamadığında ve aşağı yukarı aynı seviyede kaldığında, yakınsamıştır. Gradyan inişinin yakınsaması gereken yineleme sayısı bazen çok değişebilir. 50 yineleme, 60.000 ve hatta belki 3 milyon sürebilir, bu da yakınsama yineleme sayısını önceden tahmin etmeyi zorlaştırır.

Gradyan inişinin yakınsadığını size otomatik olarak söyleyebilen bazı algoritmalar vardır, ancak yakınsama için önceden bir eşik tanımlamanız gerekir ki bu da tahmin etmesi oldukça zordur. Bu nedenle basit grafikler tercih edilen yakınsama testidir.

Grafikler aracılığıyla gradyan inişini izlemenin bir diğer avantajı da, örneğin maliyet fonksiyonunun artması gibi düzgün çalışmadığında kolayca tespit etmemizi sağlamasıdır. Gradyan iniş kullanılırken artan bir maliyet fonksiyonunun nedeni çoğu zaman çok yüksek bir öğrenme oranıdır. 

Grafik, öğrenme eğrisinin gerçekten daha düşük bir noktaya ulaşmadan sadece yukarı ve aşağı gittiğini gösteriyorsa, öğrenme oranını düşürmeyi deneyin. Ayrıca, belirli bir problemde gradyan inişiyle başlarken, öğrenme oranları olarak 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1 vb. deneyin ve hangisinin en iyi performansı gösterdiğine bakın.

### Gradyan İniş Türleri
Temel olarak kullandıkları veri miktarında farklılık gösteren üç popüler gradyan iniş türü vardır:  

* **TOPLU GRADYAN İNİŞİ**

    Normal gradyan iniş olarak da adlandırılan toplu gradyan iniş, eğitim veri kümesindeki her örnek için hatayı hesaplar, ancak yalnızca tüm eğitim örnekleri değerlendirildikten sonra model güncellenir. Tüm bu süreç bir döngü gibidir ve buna eğitim dönemi denir.
    Toplu gradyan inişinin bazı avantajları, hesaplama verimliliğidir: kararlı bir hata gradyanı ve kararlı bir yakınsama üretir. Bazı dezavantajlar, kararlı hata gradyanının bazen modelin elde edebileceği en iyi durum olmayan bir yakınsama durumuna yol açabilmesidir. Ayrıca, tüm eğitim veri setinin bellekte olmasını ve algoritma tarafından kullanılabilir olmasını gerektirir.

 

* **STOKASTİK GRADYAN İNİŞİ**
    
    Aksine, stokastik gradyan iniş (SGD) bunu veri kümesindeki her eğitim örneği için yapar, yani her eğitim örneği için parametreleri birer birer günceller. Soruna bağlı olarak bu, SGD'yi toplu gradyan inişinden daha hızlı hale getirebilir. Bir avantaj, sık yapılan güncellemelerin oldukça ayrıntılı bir iyileştirme oranına sahip olmamıza izin vermesidir.
    Bununla birlikte, sık güncellemeler, toplu gradyan iniş yaklaşımından hesaplama açısından daha pahalıdır. Ek olarak, bu güncellemelerin sıklığı gürültülü gradyanlara neden olabilir ve bu da hata oranının yavaşça düşmek yerine sıçramasına neden olabilir.

 

* **MİNİ TOPLU DEGRADE İNİŞ**
   
    Mini toplu gradyan iniş, SGD ve toplu gradyan iniş kavramlarının bir kombinasyonu olduğu için başvurulacak yöntemdir. Eğitim veri setini küçük gruplara ayırır ve bu partilerin her biri için bir güncelleme gerçekleştirir. Bu, stokastik gradyan inişinin sağlamlığı ile toplu gradyan inişinin verimliliği arasında bir denge oluşturur.
    Genel mini parti boyutları 50 ile 256 arasında değişir, ancak diğer herhangi bir makine öğrenimi tekniğinde olduğu gibi, farklı uygulamalar için değişiklik gösterdiğinden net bir kural yoktur. Bu, bir sinir ağını eğitirken başvurulacak algoritmadır ve derin öğrenmede en yaygın gradyan iniş türüdür.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.01
    
    plt.scatter(x,y,color='red',marker='+',linewidth=5)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr   # y = mx + b
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])   # MSE
        
        plt.plot(x,y_predicted,color='green')
        
        md = -(2/n) * sum( x * (y - y_predicted))  # m türev hesabı
        bd = -(2/n) * sum(y - y_predicted)         # b türev hesabı
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m:{},\t\t b:{},\t\t cost:{},\t\t iteration:{}".format(m_curr, b_curr, cost, i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
