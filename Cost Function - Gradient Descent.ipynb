
Doğrusal Regresyon modeli, modele uyması için düz bir çizgi kullanır.



> **y = β0 + β1X + ϵ**
> * β0:Doğrunun y-eksenini kestiği yer ve regresyon sabitidir.
> * β1:Doğrunun eğimi veya regresyon katsayısıdır.
> * ϵ: Rastgele(Tesadüfi-Şans) hata değeridir.

İlk başta, değişkenler uygun şekilde optimize edilmezse, modele tam olarak uymayabilecek bir çizgi elde edersiniz. Modelin değerlerini optimize ettikçe, bazı değişkenler için mükemmel uyumu elde edebiliriz. Mükemmel uyum, gürültüyü ve aykırı değerleri göz ardı ederken veri noktalarının çoğundan geçen düz bir çizgi olacaktır. Uygun bir Doğrusal Regresyon modeli aşağıda gösterildiği gibi görünür:



Doğrusal regresyon modeli için, maliyet fonksiyonu, tahmin edilen değerlerin gerçek değerlerden çıkarılmasıyla elde edilen, modelin Ortalama Karesel Hata Kökünün minimumu olacaktır. Maliyet fonksiyonu, bu hata değerlerinin minimumu olacaktır.



Gradyan descent, belirli bir fonksiyonun yerel minimum/maksimum değerini bulmak için kullanılan yinelemeli birinci dereceden bir optimizasyon algoritmasıdır. Tüm optimize edicilerin amacı, maliyet fonksiyonunun mümkün olan en az değere ulaştığı küresel minimuma ulaşmaktır.Türevlenebilir ve dışbükey olmak üzere iki özel gereksinimi vardır.

Bir fonksiyon türevlenebilir ise, etki alanındaki her nokta için bir türevi vardır.



Dışbükey olması demek tek değişkenli bir fonksiyon için iki fonksiyonun noktalarını birleştiren doğru parçasının eğrisinin üzerinde olduğu anlamına gelir (onu kesmez).



Fonksiyondaki tüm değişkenlerin birinci dereceden türevlerinin toplamı bize gradyanı verir. Fonksiyon çıktılarının döngüsel (iterative) olarak parametrelerinin optimize edilerek yukarıdan aşağıya doğru inecek şekilde hatayı minimize etme sürecinden dolayı Azaltma/İniş (descent/descending) olarak tanımlanır.

Gradient descent’i gerçek bir problemmiş gibi düşünecek olursak;

Bir dağın zirvesinden aşağı deniz seviyesine inmeniz gerektiğini hayal edin fakat görme yeteneğiniz çok az ve etraf da karanlık bu durumda aşağı nasıl inerdiniz?

Öncelikle küçük adımlarla ilerlemeye gayret gösteririz, çok hızlı inmek tehlikeli olabilir ayağınız kayıp düşebilirsiniz amacınız genellikle küçük adımlarla inmek olur. Gradient Descent’te ise durum buna çok benzerdir, bir problemin çözümü için en uygun noktayı bulmak bunun içinse küçük adımlarla ilerleyerek hedefe varmaktır. Bu örnekte attığınız adım sayısı iterasyonun sayısına denk gelir.

Adımların büyüklüğü gradient descentte ki learning rate’e yani öğrenme oranına denk gelmektedir. Dağdan aşağıya doğru indiğiniz sırada belki 10-15 adım sonrasında daha rahat adım atmaya başladınız bunun nedeni artık bir sonraki adımınız için bir tahminleme yeteneği kazandınız 10 adımda az çok öğrenme gerçekleşti aslında. İşte gradient descentte de öğrenme işlemi bu şekilde gerçekleşmektedir. Birkaç adım iterasyona girdikten sonra öğrenme ve optimizasyon işlemleri yapılmaya başlanmaktadır.

Gradyan iniş tanımı gereği, hatanın sürekli olarak azaldığı yönü bulmanız gerekir. Bu, hatalar arasındaki farkı bularak yapılabilir. Hatalar arasındaki küçük fark, maliyet fonksiyonunun farklılaştırılması ve yokuş aşağı hareket etmek için önceki gradyan inişinden çıkarılmasıyla elde edilebilir. 



Yukarıdaki denklemlerde alfa, öğrenme oranı (**Learning Rate**) olarak bilinir. Yokuş aşağı ne kadar hızlı ineceğinize karar verir. Alfa büyükse büyük adımlar atılır, küçükse; küçük adımlar atıyorsun. Alfa çok büyükse, en küçük hata noktasını tamamen gözden kaçırabilirsiniz ve sonuçlarımız doğru olmayacaktır. Çok küçükse, modeli optimize etmek çok uzun sürer ve ayrıca hesaplama gücünü de boşa harcarsınız. Bu nedenle, optimal bir alfa değeri seçmeniz gerekir.   



### Gradyan İniş Türleri

Temel olarak kullandıkları veri miktarında farklılık gösteren üç popüler gradyan iniş türü vardır:

#### Batch Gradient Descent
Toplu gradyan inişinde, gradyan inişinin her yinelemesinde maliyet fonksiyonunun gradyanını hesaplamak için tüm veri kümesini kullanırız ve ardından ağırlıkları güncelleriz. Tüm veri kümesini hesaplamada kullandığımızdan dolayı gradyanın yakınsaması yavaştır.

#### Stochastic Gradient Descent

Rastgele gradyan iniş algoritmasında her bir yineleme içerisinde rastgele bir örnek seçilir, yitim değeri hesaplanır ve optimizasyon yapılır. Eğitim kümesindeki her bir örnek için bu işlemler yapıldıktan sonra bir sonraki yinelemeye geçilir. Bu işlemin avantajı parametrelerdeki güncellemeler daha sık yapıldığı için model minimum noktaya daha hızlı yaklaşabilir ve her seferinde bir örneği işlediğimiz için daha az hafıza gerektirir. Dezavantajları ise yitim değerinde minimuma ulaştıktan sonra bile parametreleri güncelleyip minimum değerden uzaklaşabilir ve eğitim sürecinde öğrenme hızını kademeli olarak ayarlamamız gerekir. Ek olarak, bu güncellemelerin sıklığı gürültülü gradyanlara neden olabilir ve bu da hata oranının yavaşça düşmek yerine sıçramasına neden olabilir.

#### Mini-batch Gradient Descent

Diğer iki gradyan iniş türüne göre en çok tercih edilenidir. Bu algoritmada eğitim kümesi belirli parçalara bölünür. Örnek olarak 128 adet örneği bulunan bir eğitim kümesi 16’ lık 8 adet parçaya bölünebilir. Her bir parça için modelin yitim değeri hesaplandıktan optimizasyon işlemi gerçekleştirilir. Eğitimden önce parça sayısına kara vermemiz gerekir. Genellikle 2’ nin katları tercih edilir 2,4,8,16,32,64 olacak şekilde.





Maliyet işlevi, bir veri kümesi için bir makine öğrenimi modelinin performansını ölçer. Maliyet işlevi, tahmin edilen ve beklenen değerler arasındaki hatayı ölçer ve bu hatayı tek bir gerçek sayı biçiminde sunar. Probleme bağlı olarak, maliyet fonksiyonu birçok farklı şekilde oluşturulabilir. Maliyet fonksiyonunun amacı, minimize veya maksimize etmektir. Model parametrelerini optimize etmek için gradyan inişine dayanan algoritmalar için her fonksiyonun türevlenebilir olması gerekir.

    Mavi noktalar kedileri, kırmızı noktalar ise köpekleri temsil ediyor. Aşağıda, yukarıdaki sınıflandırma probleminin bazı çözümleri bulunmaktadır.

    

    Esasen her üç sınıflandırıcı da çok yüksek doğruluğa sahiptir, ancak üçüncü çözüm en iyisidir çünkü herhangi bir noktayı yanlış sınıflandırmaz. Tüm noktaları mükemmel bir şekilde sınıflandırmasının nedeni, çizginin neredeyse tam olarak iki grup arasında olması ve herhangi bir gruba daha yakın olmamasıdır. İşte burada maliyet fonksiyonu kavramı devreye giriyor. Maliyet fonksiyonu optimum çözüme ulaşmamıza yardımcı oluyor. Maliyet fonksiyonu, “algoritmamızın/modelimizin performansını” değerlendirme tekniğidir.

    Hem model tarafından tahmin edilen çıktıları hem de gerçek çıktıları alır ve modelin tahmininde ne kadar yanlış olduğunu hesaplar. Tahminlerimiz gerçek değerlerden çok farklıysa daha yüksek bir sayı verir. Tahminleri iyileştirmek için modelimizi ayarlarken, maliyet işlevi, modelin nasıl geliştiğinin bir göstergesi olarak işlev görür. Bu aslında bir optimizasyon problemidir. Optimizasyon stratejileri her zaman “maliyet fonksiyonunu en aza indirmeyi” amaçlar.

### Maliyet fonksiyonunun türleri
Makine öğreniminde birçok maliyet fonksiyonu vardır ve her birinin, bunun bir regresyon problemi mi yoksa sınıflandırma problemi mi olduğuna bağlı olarak kullanım durumları vardır.
> **Regresyon için olan ölçütler:**
> * L1 Loss /Ortalama Mutlak Hata (MAE)
> * L2 Loss / Hata Karelerinin Ortalaması (MSE)
> * Root Mean Squared Error (RMSE)


> **Sınıflandırma için loss fonksiyonları:**
> * Log Loss (Cross Entropy Loss)
> * SVM Loss (Hinge Loss)


* **Ortalama mutlak hata (MAE)**, veri kümesindeki gerçek ve tahmin edilen değerler arasındaki mutlak farkın ortalamasını temsil eder. Veri setindeki artıkların ortalamasını ölçer. Mutlak hata değerinin toplamını alır, hata terimlerinin toplamının daha doğrudan bir temsilidir.



* **Ortalama Karesel Hata (MSE)**, veri setindeki orijinal ve tahmin edilen değerler arasındaki farkın karesinin ortalamasını temsil eder. Kalıntıların varyansını ölçer. Ortalama Kare Hatası tahmin edilen sonuçlarınızın gerçek sayıdan ne kadar farklı olduğuna dair size mutlak bir sayı verir. Tek bir sonuçtan çok fazla içgörü yorumlayamazsınız, ancak size diğer model sonuçlarıyla karşılaştırmak için gerçek bir sayı verir ve en iyi regresyon modelini seçmenize yardımcı olur.



* **Kök Ortalama Kare Hatası (RMSE)**, Ortalama Kare hatasının kareköküdür. Kalıntıların standart sapmasını ölçer. MSE’den daha sık kullanılır çünkü bazen MSE değeri kolayca karşılaştırılamayacak kadar büyük olabilir. Bu yüzden MSE hata karesi ile hesaplanır ve böylece yorumlamayı kolaylaştırır. Fakat MSE aykırı değerlere karşı çok duyarlıdır.



* **Belirleme katsayısı veya R-kare**, modeldeki bağımsız değişkenlere göre bağımlı değişkenin varyasyon oranını yani bağımlı değişkendeki değişkenliğin ne kadarının model tarafından açıklanabileceğini ölçer. Korelasyon katsayısının karesidir. R Kare aşırı uyum (overfitting) sorununu dikkate almaz. Regresyon modelinin çok fazla bağımsız değişkeni varsa model eğitim verilerine çok iyi uyabilir ama testte istenen başarıyı gösteremeyebilir. Bu nedenle Düzeltilmiş R Kare kullanılır. Düzeltilmiş R Kare modele eklenen ek bağımsız değişkenleri cezalandırır ve aşırı uyum sorununu çözer.



* **Düzeltilmiş R kare**, R karenin değiştirilmiş bir versiyonudur ve modeldeki bağımsız değişkenlerin sayısına göre ayarlanır ve her zaman R²'den küçük veya ona eşit olacaktır. Aşağıdaki formülde n , verilerdeki gözlem sayısıdır. ve k , verilerdeki bağımsız değişkenlerin sayısıdır.



### Maliyet Fonsiyonu ve Gradient Descent

Bir maliyet fonksiyonu, en aza indirmek istediğiniz bir şeydir. Örneğin, maliyet işleviniz, eğitim kümenizdeki hataların karelerinin toplamı olabilir. Gradyan iniş, çoklu değişkenli bir fonksiyonun minimumunu bulmak için kullanılan bir yöntemdir. Böylece, maliyet fonksiyonunuzu en aza indirmek için gradyan inişini kullanabilirsiniz. Maliyetiniz K değişkeninin bir fonksiyonuysa, gradyan, maliyetin en hızlı arttığı yönü tanımlayan uzunluk-K vektörüdür. Yani gradyan inişinde, gradyanın negatifini maliyetin minimum olduğu noktaya kadar takip edersiniz. Birisi bir makine öğrenimi bağlamında gradyan inişinden bahsediyorsa, muhtemelen maliyet işlevi ima edilir (gradyan iniş algoritmasını uyguladığınız işlevdir).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.01
    
    plt.scatter(x,y,color='red',marker='+',linewidth=5)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr   # y = mx + b
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])   # MSE
        
        plt.plot(x,y_predicted,color='green')
        
        md = -(2/n) * sum( x * (y - y_predicted))  # m türev hesabı
        bd = -(2/n) * sum(y - y_predicted)         # b türev hesabı
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m:{},\t\t b:{},\t\t cost:{},\t\t iteration:{}".format(m_curr, b_curr, cost, i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
