### Maliyet Fonsiyonu ve Gradient Descent

Bir maliyet fonksiyonu, en aza indirmek istediiniz bir eydir. rnein, maliyet ileviniz, eitim kmenizdeki hatalarn karelerinin toplam olabilir. Gradyan ini, oklu deikenli bir fonksiyonun minimumunu bulmak iin kullanlan bir yntemdir. Bylece, maliyet fonksiyonunuzu en aza indirmek iin gradyan iniini kullanabilirsiniz. Maliyetiniz K deikeninin bir fonksiyonuysa, gradyan, maliyetin en hzl artt yn tanmlayan uzunluk-K vektrdr. Yani gradyan iniinde, gradyann negatifini maliyetin minimum olduu noktaya kadar takip edersiniz. Birisi bir makine renimi balamnda gradyan iniinden bahsediyorsa, muhtemelen maliyet ilevi ima edilir (gradyan ini algoritmasn uyguladnz ilevdir).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 100
    n = len(x)
    learning_rate = 0.01
    
    plt.scatter(x,y,color='red',marker='+',linewidth=5)
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr   # y = mx + b
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])   # MSE
        
        plt.plot(x,y_predicted,color='green')
        
        md = -(2/n) * sum( x * (y - y_predicted))  # m trev hesab
        bd = -(2/n) * sum(y - y_predicted)         # b trev hesab
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print("m:{},\t\t b:{},\t\t cost:{},\t\t iteration:{}".format(m_curr, b_curr, cost, i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
