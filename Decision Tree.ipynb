## Sklearn'de DecisionTreeClassifier' Anlamak

> DecisionTreeClassifierSklearn'de snfta bulunan parametreler

```python
DecisionTreeClassifier(
    criterion = 'gini', # Bir blmenin kalitesini lme ilevi. 'gini' - 'entropy'.
    splitter = 'best', # En iyi blnmeyi seme stratejisi. 'best' - 'random'
    max_depth = None, # Aacn maksimum derinlii. none ise, dmler tm yapraklar saf olana veya min_samples_split'ten daha azn ierene kadar geniletilir.
    min_samples_split = 2, # Bir dm blmek iin gereken minimum rnek says.
    min_samples_leaf = 1, # Bir yaprak dmde olmas gereken minimum rnek says.
    min_weight_fraction_leaf = 0.0, # Bir dmde olmas gereken tm girdi rneklerinin arlklarnn toplamnn minimum arlkl fraksiyonu.
    max_features = None, # En iyi ayrm ararken dikkate alnmas gereken zelliklerin says. "int-float-'auto'-'sqrt'-'log2'-none" olabilir.
    random_state = None, # Tahmin edicinin rastgelelii iin kontrol
    max_leaf_nodes=None, # Maksimum sayda dm ieren bir aa bytn. Yok ise, snrsz sayda mmkndr.
    min_impurity_decrease=0.0, # Bu blme, safszl bu deerden byk veya bu deere eit azaltrsa, bir dm blnecektir.
    class_weight=None, # Farkl snflarla ilikili arlklar.
    ccp_alpha=0.0 # Minimum Maliyet-Karmaklk Budamas iin kullanlan karmaklk parametresi.
)
```

# Sklearn'de Karar Aac Snflandrclar iin Hiperparametre Ayar

Bu reticiyi bitirmek iin, baz hiper parametrelerini ayarlayarak modelimizin doruluunu nasl gelitirebileceimize bir gz atalm. Hiper parametreler, bir makine renimi modeli olutururken belirttiiniz deikenlerdir. Bu, rnein, algoritmann verileri nasl bldn (entropi veya gini safszl ile) ierir.

Hiper parametre ayar, daha sonra, daha yksek bir doruluk puan salamak iin bu deerleri ayarlama srecini ifade eder. Bunu yapmann bir yolu, basite, farkl deerleri girmek ve hangi hiper parametrelerin en yksek puan dndrdn grmektir.

Scikit-learn, bir dizi farkl kombinasyonu denemek yerine, bu testleri otomatikletirmenin bir yolunu sunar. Bu yntem, ilemi nemli lde daha hzl hale getiren GridSearchCV yntemidir. Denemek iin farkl deerlerden oluan bir szlk salamamz yeterlidir ve Scikit-Learn sreci bizim yerimize halleder.

Ayrca, snf bir apraz dorulama(cross-validation) srecini tamamlar. Bu, snfn fazla uydurmay nlemeye yardmc olmak iin farkl eitim ve test verileri kombinasyonlar arasnda gei yapaca anlamna gelir. rnein, %20'lik bir test boyutu belirlediimizde, apraz dorulama btne gre bu %20'nin farkl blmleri arasnda gei yapacaktr.

> Hem en iyi hiperparametreleri bulmak hem de ayn anda apraz dorulama uygulamak iin GridSearchCV snfn nasl uygulayabileceimizi grelim.

Bunu yapabilmek iin ncelikle hangi hiperparametreleri test edeceimize karar vermemiz gerekiyor.

```python
from sklearn.model_selection import GridSearchCV

params = {
    'criterion':  ['gini', 'entropy'],  # bir blmenin kalitesini belirlemek iin kullanlan ilev
    'max_depth':  [None, 2, 4, 6, 8, 10], # aacn maksimum derinlii
    'max_features': [None, 'sqrt', 'log2', 0.2, 0.4, 0.6, 0.8], # blme yaparken dikkate alnmas gereken maksimum zellik says
    'splitter': ['best', 'random'] # her dmde blmeyi semek iin kullanlan strateji
}

clf = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=params,
    cv=5,
    n_jobs=5,
    verbose=1,
)

clf.fit(X_train, y_train)
print(clf.best_params_)
```
