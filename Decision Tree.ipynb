## Karar Ağaçları İle Sınıflandırma
Karar ağaçları, verileri yüksek doğruluk dereceleriyle sınıflandırmanıza izin veren sezgisel, denetimli bir makine öğrenimi algoritmasıdır. Karar Ağaçları, karar verirken genellikle insanın düşünme yeteneğini taklit eder ve ağaç benzeri bir yapı gösterdiği için anlaşılması kolaydır.

Bir pazarlama yöneticisi olarak, ürününüzü satın alma olasılığı en yüksek olan bir dizi müşteri istiyorsunuz. Kitlenizi bularak pazarlama bütçenizden bu şekilde tasarruf edebilirsiniz. Bir kredi yöneticisi olarak, daha düşük bir kredi temerrüt oranı elde etmek için riskli kredi başvurularını belirlemeniz gerekir. Müşterileri potansiyel ve potansiyel olmayan müşteriler veya güvenli veya riskli kredi başvuruları grubuna ayırma işlemi, bir sınıflandırma problemi olarak bilinir. Sınıflandırma, **öğrenme adımı** ve **tahmin adımı** olmak üzere iki aşamalı bir süreçtir. *Öğrenme adımında*, verilen eğitim verilerine dayalı olarak model geliştirilir. *Tahmin adımında*, model verilen veri için yanıtı tahmin etmek için kullanılır. Karar Ağacı, anlaşılması ve yorumlanması en kolay ve popüler sınıflandırma algoritmalarından biridir. Hem sınıflandırma hem de regresyon türü problemler için kullanılabilir.

Karar Ağacı, beyaz kutu tipi bir makine öğrenimi algoritmasıdır. Neural Network gibi kara kutu tipi algoritmalarda bulunmayan dahili karar verme mantığını paylaşır. Eğitim süresi sinir ağı algoritmasına göre daha hızlıdır. Karar ağaçlarının zaman karmaşıklığı, verilen verideki kayıt sayısı ve öznitelik sayısının bir fonksiyonudur. Karar ağacı, olasılık dağılım varsayımlarına bağlı olmayan, dağılımdan bağımsız veya parametrik olmayan bir yöntemdir. Karar ağaçları, yüksek boyutlu verileri iyi bir doğrulukla işleyebilir.



### Karar Ağaçlarında Dallanma Kriterleri
Karar ağaçlarında en önemli sorunlardan birisi herhangi bir kökten itibaren dallanmanın hangi kıstasa göre yapılacağıdır.Her farklı kriter için bir karar ağacı algoritması karşılık gelmektedir. 

Algoritmaları gruplandıracak olursak;
* Entropiye Dayalı Algoritmalar
* Sınıflandırma ve Regresyon ağaçları
* Bellek tabanlı sınıflandırma algoritmaları

## Karar Ağacı Algoritması Nasıl Çalışır?
Herhangi bir karar ağacı algoritmasının arkasındaki temel fikir şu şekildedir:

1. Ağaca, tüm veri setini içeren kök düğümle başlayın, diyor S.
2. Nitelik Seçim Ölçütünü (ASM) kullanarak veri kümesindeki en iyi özelliği bulun .
3. S'yi en iyi nitelikler için olası değerleri içeren alt kümelere bölün.
4. En iyi özelliği içeren karar ağacı düğümünü oluşturun.
5. Adım -3'te oluşturulan veri setinin alt kümelerini kullanarak yinelemeli olarak yeni karar ağaçları oluşturun. Düğümleri daha fazla sınıflandıramayacağınız ve son düğümü yaprak düğüm olarak adlandıracağınız bir aşamaya gelene kadar bu işleme devam edin.
    
## Nitelik Seçim Ölçüleri
Nitelik seçim ölçüsü, verileri mümkün olan en iyi şekilde bölümleyen bölme kriterini seçmek için bir buluşsal yöntemdir. Belirli bir düğümdeki demetler için kesme noktalarını belirlememize yardımcı olduğu için bölme kuralları olarak da bilinir. ASM, verilen veri setini açıklayarak her özelliğe (veya özniteliğe) bir sıralama sağlar. En iyi puan özelliği, bir bölme özelliği olarak seçilecektir . Sürekli değerli bir öznitelik olması durumunda, dallar için ayrık noktaların da tanımlanması gerekir. En popüler seçim ölçütleri;
* Bilgi Kazanımı 
* Kazanç Oranı
* Gini Endeksi

### Bilgi Kazanımı 
Bilgi kazancı, bir özniteliğe dayalı olarak bir veri kümesinin bölümlenmesinden sonra entropideki değişikliklerin ölçümüdür. Bir özelliğin bize bir sınıf hakkında ne kadar bilgi sağladığını hesaplar. Bilgi kazancının değerine göre düğümü bölüp karar ağacını oluşturuyoruz.Bir karar ağacı algoritması her zaman bilgi kazancının değerini maksimize etmeye çalışır ve en yüksek bilgi kazancına sahip bir düğüm/öznitelik önce bölünür.

Shannon, girdi setinin safsızlığını ölçen entropi kavramını icat etmiştir. **Fizik ve matematikte entropi, sistemdeki rastgelelik veya kirlilik olarak adlandırılır. Bilgi teorisinde, bir grup örnekteki safsızlığı ifade eder.** Bilgi kazancı entropinin azalmasıdır. Bilgi kazancı, verilen öznitelik değerlerine dayalı olarak veri kümesinin bölünmesinden önceki entropi ile bölünmesinden sonraki ortalama entropi arasındaki farkı hesaplar. ID3 (Iterative Dichotomiser) karar ağacı algoritması bilgi kazancını kullanır.

<div class="alert alert-block alert-success">
<b>ID3 Algoritması :</b>

Hesaplanan entropi değerleri, genel entropiden çıkartılarak; hangi özniteliğin entropi değerini en çok azalttığı tespit edilir.

### Budama: Optimal Karar Ağacı Elde Etme
Budama, optimal karar ağacını elde etmek için gereksiz düğümleri bir ağaçtan silme işlemidir.

Çok büyük bir ağaç, fazla uydurma riskini artırır ve küçük bir ağaç, veri kümesinin tüm önemli özelliklerini yakalayamayabilir. Bu nedenle, doğruluğu azaltmadan öğrenme ağacının boyutunu küçülten bir teknik Budama olarak bilinir. Kullanılan başlıca iki tür ağaç budama teknolojisi vardır:

* Maliyet Karmaşıklığı Budama
* Azaltılmış Hata Budama.

#### Karar Ağacının Avantajları
* Bir insanın gerçek hayatta herhangi bir karar verirken izlediği süreci takip ettiği için anlaşılması kolaydır.
* Kararla ilgili sorunları çözmek için çok yararlı olabilir.
* Bir problem için tüm olası sonuçları düşünmeye yardımcı olur.
* Diğer algoritmalara göre daha az veri temizleme gereksinimi vardır.
* Veri sağlamlığı algoritma her türlü veriyi güzel bir şekilde işler.
* Yorumlanabilirlik karar ağacını görselleştirebilirsiniz. Bu büyük bir avantaj çünkü çoğu algoritma kara kutu gibi çalışıyor ve algoritmanın belirli bir sonucu tahmin etmesini sağlayan şeyin ne olduğunu net bir şekilde saptamak zor.

#### Karar Ağacının Dezavantajları

* Karar ağacı, onu karmaşık hale getiren çok sayıda katman içerir.
